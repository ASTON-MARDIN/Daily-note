# 哈希表


## 怎么理解 ？

1. 索引  

数组有下标的索引，哈希表的存储结构虽然是数组，但是下标可以理解为是 经过哈希函数得到的哈希值

假设 哈希函数 hash() 是简单的  取模运算（%)

    有如下三个数据  （32，test）  （123，stream）  （64， cpp）  和一个 长度为8的数组
   
    那么第一个数据 （32，test）   的在数组中的下标索引值 为 32%8 =  0；
   
    而且最重要的是，当我们想得到 此数据的 value 时候，只需要向哈希函数 带入key 得到索引值 取出数据就完成了

    时间复杂度 即为 O（1）

    这就是哈希表精妙的地方

2. 

**任意数对2的N次方取模时，等同于其和2的N次方-1作位于运算。**

公式表述为：

>k % 2^n = k & (2^n - 1)

而位于运算相比于取模运算速度大幅度提升（按照Bruce Eckel给出的数据，大约可以提升5～8倍）。

3. 负载因子

负载因子是哈希表的重要参数，其定义为：哈希表中已存有的元素与哈希表长度的比值。

它是一个浮点数，表示哈希表目前的装满程度。由于表长是定值，而表中元素的个数越大，表中空余位置就会更少，发生碰撞的可能性也会进一步增大。

哈希表的扩容策略依赖于负载因子阈值。基于性能与空间的选择，JDK标准库将HashMap的负载因子阈值定为**0.75**。

## 怎么用？

包含在C++  **map** 库中 **可以直接调用** ，不需要直接现场实现hash

1. 添加key value  

 > map.insert()

2. 查找 key

 > map.find()
